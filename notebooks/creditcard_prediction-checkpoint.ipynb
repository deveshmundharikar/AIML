```python
x=df.drop(columns="Class",axis=1)
y=df["Class"]
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape) 
```

```python
print(x_train.columns)
print(x_test.columns)
```

```python
print("this is y_train data :---",y_train) 
print("this is y_test data :----",y_test)
```

```python
from sklearn.preprocessing import RobustScaler 
```

```python
rs=RobustScaler()
x_train_rs=rs.fit_transform(x_train)
x_test_rs=rs.transform(x_test) 
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=50,max_iter=100,n_jobs=-1) 
lr.fit(x_train_rs,y_train)
print(lr.predict(x_test_rs))  
print(lr.score(x_test_rs,y_test))  
print("Iterations:", lr.n_iter_)
```

```python
from sklearn.metrics import classification_report
print( classification_report(y_train, lr.predict(x_train_rs), target_names=['Not Fraud', 'Fraud']))
print( classification_report(y_train, lr.predict(x_train), target_names=['Not Fraud', 'Fraud']))
print( classification_report(y_test, lr.predict(x_test_rs), target_names=['Not Fraud', 'Fraud']))
print( classification_report(y_test, lr.predict(x_test), target_names=['Not Fraud', 'Fraud']))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_prob = lr.predict_proba(x_test_rs)[:,1]  
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob):.2f}")
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
score = lr.score(x_test_rs, y_test)

plt.bar(["Accuracy"], [score], color="green")
plt.ylim(0,1)
plt.title("Model Accuracy")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```

```python

from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, precision_recall_curve

# Predict values
y_pred = lr.predict(x_test_rs)
y_prob = lr.predict_proba(x_test_rs)[:, 1]  # probability for class 1

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Non-Fraud (0)", "Fraud (1)"],
            yticklabels=["Non-Fraud (0)", "Fraud (1)"],
            ax=axes[0,0])
axes[0,0].set_title("Confusion Matrix")
axes[0,0].set_xlabel("Predicted")
axes[0,0].set_ylabel("Actual")

# 2. Accuracy Score (Bar Graph)
score = lr.score(x_test_rs, y_test)
axes[0,1].bar(["Accuracy"], [score], color="green")
axes[0,1].set_ylim(0,1)
axes[0,1].set_title(f"Model Accuracy = {score:.2f}")

# 3. ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
axes[1,0].plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob):.2f}")
axes[1,0].plot([0,1],[0,1],'--', color='gray')
axes[1,0].set_xlabel("False Positive Rate")
axes[1,0].set_ylabel("True Positive Rate")
axes[1,0].set_title("ROC Curve")
axes[1,0].legend()

# 4. Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
axes[1,1].plot(recall, precision)
axes[1,1].set_xlabel("Recall")
axes[1,1].set_ylabel("Precision")
axes[1,1].set_title("Precision-Recall Curve")

plt.tight_layout()
plt.show()

```

```python
# Custom loop for iterations
from sklearn.metrics import log_loss

losses = []

for i in range(1, 50):   # 50 iterations tak try karenge
    lr.max_iter = i
    lr.fit(x_train_rs, y_train)

    # Probability predictions
    y_prob = lr.predict_proba(x_train_rs)[:, 1]

    # Calculate log loss (convergence check)
    ll = log_loss(y_train, y_prob)
    losses.append(ll)

# Plot convergence curve
plt.plot(range(1, len(losses)+1), losses, marker="o")
plt.xlabel("Iterations")
plt.ylabel("Log Loss")
plt.title("Logistic Regression Convergence Curve")
plt.show() 
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state= 42, n_jobs= -1)

rf.fit(x_train_rs, y_train)
print(rf.predict(x_test_rs))
print(rf.score(x_test_rs,y_test))
 
print('\nTrain rf :\n', classification_report(y_train, rf.predict(x_train_rs),target_names=['Not Fraud', 'Fraud']))
print('\nTest rf :\n', classification_report(y_test, rf.predict(x_test_rs),target_names=['Not Fraud', 'Fraud']))
```
