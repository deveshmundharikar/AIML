```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
import seaborn as sns 
```

```python
x,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
```

```python
x.shape 
```

```python
plt.scatter(x[:,0],x[:,1])  
plt.show()  
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split( x,y, test_size=0.2,random_state=50) 
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape) 
 
```

```python
from sklearn.cluster import KMeans  
```

```python
#lbow method
wcss=[]
for i in range(1,11) :
    means=KMeans( n_clusters=i,init='k-means++')
    means.fit(x_train)
    wcss.append(means.inertia_)  
```

```python
wcss 
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.show()  
```

```python
kmeans=KMeans(  n_clusters=3,init='k-means++')
 
```

```python
y_labels=kmeans.fit_predict(x_train )  
```

```python
y_test=kmeans.predict(x_test )   
plt.scatter(x_test[:,0],x_test[:,1],c=y_test)     
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_labels)  
```

```python
#using second method 
```

```python
#using nee locator 
# ! pip install kneed
from kneed import KneeLocator  
```

```python
k1=KneeLocator(range(1,11),wcss,curve = 'convex',direction = 'decreasing')
```

```python
k1.elbow 
```

```python
#perfomace metrics
#silhote score 
from sklearn.metrics import silhouette_score   
```

```python
sil_coefi=[]
for k in range (2,11):
    kmeans =KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    score=silhouette_score(x_train,kmeans.labels_)
    sil_coefi.append(score) 
```

```python
sil_coefi 
```

```python
#output zero 
```

```python
plt.plot(range(1,11),wcss,sil_coefi)
plt.xticks(range(1,11))
plt.show()  
```

```python




#hierarichal clustering implementation
import pandas as pd 
import numpy as np
import seaborn as sns 
from sklearn import datasets
import matplotlib.pyplot as plt   
```

```python
df=sns.load_dataset("iris") 
```

```python
df.head()  
```

```python
df1=df.drop(columns="species",axis=1)  
```

```python
df1.shape  
```

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split 
```

```python
sc=StandardScaler()  
standard_df1=sc.fit_transform(df1)  
```

```python
from sklearn.decomposition import PCA 
```

```python
pca=PCA(n_components=2)
 
```

```python
pca 
```

```python
p1=pca.fit_transform(standard_df1)  
```

```python
plt.scatter(p1[:,0],p1[:,1]) 
```

```python
#using scipy for dendogram 
import scipy.cluster.hierarchy as sc 
#ploting dendogram 
plt.figure(figsize=(20,17))
plt.title("dendogram")
#create dendogram


#fix line imp line for create dendogram 
sc.dendrogram(sc.linkage(p1,method="ward"))
plt.title("dendogram")   
```

```python
#clustering
from sklearn.cluster import AgglomerativeClustering
 
```

```python
cluster=AgglomerativeClustering( n_clusters=2,linkage='ward')
cluster.fit(p1) 
```
