```python
from sklearn.pipeline import Pipeline 
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer  
from sklearn.preprocessing import OneHotEncoder   
```

```python
imput_mean=["'AGE'"] 
imput_median=['SUM_YR_1', 'SUM_YR_2']
cat_enc=[4]
```

```python
imputer_mean=Pipeline(steps=[("imputer",SimpleImputer(strategy='mean'))])
imputer_median=Pipeline(steps=[("imputer",SimpleImputer(strategy='median'))]) 
```

```python
uf=ColumnTransformer(transformers=[
(" num_mean_imputer",imputer_mean,imput_mean),
("num_median_imputer",imputer_median,imput_median),
("enc",OneHotEncoder(drop="first",sparse_output=True ,dtype="int",handle_unknown='error'),[3]) 
],remainder='passthrough')   
```
